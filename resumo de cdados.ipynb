A função que estudamos é a função densidade de probabilidade e ela é utilizada para construir um modelo teórico, que podemos aplicar a um conjunto de dados. Numa função dessas, a integral do intervalo x (variando em x - dx) tem que ser igual a 1.

A probabilidade de X ser maior do que certo valor ou menor do que certo valor é descrita pela integral - indo do primeiro limite ao segundo limite (limites do intervalo estudado ou valor pedido no enunciado) - da função densidade de probabilidade (f(x) - pode ser uniforme, ...) variando em x:



Sabemos que a integral de um ponto é igual a 0, então a probabilidade de X ser igual a um só valor (ex: X = a - estando a dentro di intervalo x) é sempre 0. Portanto ignoramos o '=' dos '<=' e '>='. Com as variáveis contínuas, apenas é possível calcular a probabilidade de X ser maior ou menor do que um valor (pq é uma parte da área (equivalente a 1) do gráfico da função densidade f(x), que descreve a distribuição de frequências relativas).

OBS: PELO MENOS significa um "dado que", então para calcular algo do tipo usamos a fórmula do "dado que" (probabilidade da intersecção entre os dois dividido pela probabilidade do dado).

Num caso em que a função é uniforme, usamos o seguinte código para calcular a frequência relativa de um X (a freq. é o y da função f(x), enquanto X é o ponto estudado), considerando X ~ (valor_minimo, valor_maximo):

# Modelo probabilístico contínuo normal:
### Aula 17

A simetria indica que os dados estão distribuídos igualmente e, portanto, a moda, a média e a mediana são todas iguais a mu. Mu (média) desloca o gráfico horizontalmente (se mantendo simétrico) e sigma (desvio padrão) modifica quão estreita curva é (considerando que se todos os valores fossem iguais, o gráfico descrito seria uma reta vertical partindo do valor comum mu).

Ponto de maior densidade é a média, que é igual à mediana --> todos os modelos normais são simétricos

Notação correta da média e desvio padrão: X ~ N(média; desvio_padrão^2) - o ~ se lê "segue a distribuição"

Temos um gráfico padrão que descreve a relação entre a média, o desvio padrão e a distribuição normal. Ele serve para todos os casos em que o valor que queremos que X seja menor ou maior (x), se tiver mu subtraído, seja múltiplo inteiro do desvio padrão e permite que facilmente calculemos uma probabilidade se tivermos a média e o desvio padrão:



Podemos interpretar esse gráfico pensando que a probabilidade de encontrar X (um valor) - considerando para o exemplo que X é maior que a média mu - menor do que x (valor entregue pelo enunciado) é a probabilidade de encontrar a média mais a probabilidade de encontrar a x menos a probabilidade de encontrar a média. 

Exemplo:
Se temos que a média é 5 e o desvio padrão é 1,5 e queremos encontrar a probabilidade de X ser menor que 6,5:

P(X<=6,5) = P(X<=5) + P(5<=X<=6,5)     ==>     olhando o gráfico: (P de ser menor que a média) 0,5 + (P de estar entre a média e a média + um desvio padrão) 0,3413 = 0,8413

Nesse mesmo caso, se quiséssemos calcular a probabilidade de encontrar um X maior do que um número do qual se subtraírmos a média, não é múltiplo inteiro do desvio padrão, vamos precisar de comandos do python:

# Teorema do limite central
### Aula 20

Na estatística usamos a inferência para gerar conhecimennto sobre mu usando a média amostral (x com risquinho em cima), que são apenas algumas informações. Quanto mais x a gente usa fazendo a média aritmética, mais próximo do espaço amostral estamos, então menor é a margem de erro que a média amostral nos permitir inferir sobre a média populacional (mu). A médial amostral (estimativa da média) também muda de acordo com os dados que pegamos (se usamos x1, x2 e x3 ou x4, x5 e x6, recebemos resultados diferentes), ou seja cada uma delas é uma variável aleatória.

$\overline{x}$ ==> são as diversas médias amostrais que podemos adquirir utilizando amostras difrentes para essa média

$\overline{X}$ ==> é uma variável aleatória (cada um dos $\overline{x}$s é um $\overline{X}$)

n ==> quantidade de $\overline{X}$s

Quanto maior o nosso n, menor fica a variância da função. 
Usamos principalmente a média da amostra $\overline{X}$, pois essa é a média amostral.

Para pegar uma amostra aleatória considerando X ~ Uniforme(valor_minimo, valor_maximo), usamos o código a seguir:

```python
amostra = stats.uniform.rvs(valor_minimo, valor_maximo - valor_minimo, size=n)
```

Se a variável aleatória X é normal, $\overline{X}$ é normal para qualquer tamanho amostral n. Porém, se X não for normal, $\overline{X}$ tende a ser normal quanto maior n é (converge a uma distribuição normal). Em ambos os casos, a variância de $\overline{X}$ equivale à variância de X divida por n.

OBS: se X é normal, $\overline{X}$ é normal de forma exata. Se X não é normal, $\overline{X}$ é normal de forma aproximada.

As hipóteses são afimações de inferência sobre alguma característica da população por meio de uma variável de interesse X (não afirmações da amostra):
- H0: A hipóteses nula é a que simula igualdade (testa se o mu novo é igual ao mu anterior) - é a hipótese colocada a teste. Ou também inclui tudo que não está incluído na hipótese alternativa.
- H1: A hipótese alternativa é quando testamos se a média nova é menor, maior ou diferente da original.

Se ao usar a hipótese nula para a fazer a inferência, a estimativa para a média populacional tem uma probabilidade muito pequena de surgir na curva normal do contexto original, tendemos a rejeitar a hipotese nula. Nesse caso, testamos a hipótese alternativa usando a curva de $\overline{x}$ e o nível de significância (que define qual é a área que define a região crítica - ou seja, área que separa as médias 'aceitáveis' das que a gente rejeita): se a média da amostra estiver fora do ponto que separa as regiões crítica (RC) da aceitável - que é o ponto crítico -, rejeitamos a hipótese nula e não rejeitamos a hipótese alternativa. Usamos o ppf para encontrar o ponto crítico (ponto C ou ponto de corte), com mu em loc e sigma/sqrt(n) (que é o desvio padrão da curva de $\overline{x}$ considerando sigma o desvio padrão da cutva normal original) como scale. A área depois do ponto C é chamada alpha.

# Regressão
### Aula 26

Num modelo de regressão, y é a variável resposta (target) e x é a variável explicativa (feature). Utilizando uma função de primeiro grau $\hat{y}$ = $\beta_0$ + $\beta_1 * x$, montamos uma reta $\hat{y}$ (reta ajustada para prever y) que indica a tendência do gráfico de dispersão estudado. A inclinação dessa reta, como indica a função, é $\beta_1$, e quanto mais distante de 0 (que representaria que o gráfico é uma reta horizontal), mais provável é que a reta seja precisa, o que significaria que a variável x é relevante para prever y (que é o objetivo da regressão) - hipótese alternativa. A hipótese nula $\beta_0$ não é rejeitada quando $\beta_1$ é 0, e representa o caso em que a variável x não é relevante para a previsão de y.

    Exemplos:
- caso em que rejeitamos a hipótese nula: estamos prevendo se a variável idade pode nos ajudar a prever a variável anos de estudo. É nítido que a idade de alguém é um fator relevante para indicar quantos anos de estudo ela tem, portanto a hipótese nula é rejeitada - $\beta_1$ é diferente de 0.
- caso em que rejeitamos a hipótese alternativa: estamos prevendo se a variável altura pode nos ajudar a prever a variável salário. É nítido que não há relação entre a altura de alguém e seu salário, portanto a hipótese alternativa é rejeitada - $\beta_1$ é igual ou muito próximo de 0.

O valor-p é a probabilidade de termos uma amostra como a estudada se H0 fosse verdade. No caso de estudo da inclinação $\beta_1$, o valor-p indica a mesma coisa: quanto mais alto é esse valor, mais alta é a chance da minha amostra ser semelhante a H0 e, portanto, mais credibilidade temos para não rejeitar H0. Então para rejeitarmos H0, o valor-p tem que ser muito pequeno, mas não precisamos que $\beta_1$ seja exatamente 0.

OBS: A distribuição t é usada quando não temos a variância populacional e temos que estimá-la (através do uso desse distribuição e da amostra). Quando temos a variância populacional, usamos distribuição normal.
